## Analysis of TTL values of IP addresses in a .pcap file

```python
import pandas as pd
import numpy as np
import sys
import subprocess
from io import StringIO
```

```python
file_name = 'a5bf72a3499a3b4d8e3f0371dbb3bdcf.pcap' #Example of manual analysis
# file_name = sys.argv[1] #For automated analysis
```

```python
command = 'tshark -n -r '+ file_name+' -E separator=; -E header=y -T fields -e ip.src -e ip.ttl'
exec_command = subprocess.Popen(command.split(), stdout=subprocess.PIPE)
output = StringIO(exec_command.communicate()[0].decode('utf-8'))
```

```python
df = pd.read_csv(output, error_bad_lines=False, sep=';')
```

```python
df.head()
```

```python
summary = df.groupby(['ip.src'])['ip.ttl'].agg(['count','mean','median', 'unique', np.ptp,]).reset_index()
summary.columns=['ip','num_pkts','mean','median','distinct_values','delta']
```

```python
summary
```

```python
summary.to_csv('ttl_'+file_name+'.csv')
```
